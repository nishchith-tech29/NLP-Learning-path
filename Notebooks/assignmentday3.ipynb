```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
with open("SMSSpamCollection", "r", encoding="utf-8") as file:
    paragraph = file.read()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
lemmatizer = WordNetLemmatizer()

# 4. Text Cleaning & Pre-processing Pipeline
corpus = []
for i in range(len(sentences)):
    # Remove special characters using Regex (Keep only a-z and A-Z)
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    
    # Convert to lowercase
    review = review.lower()
    
    # Split into words
    review = review.split()
    
    # Remove stopwords and apply Lemmatization
    review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    
    # Rejoin the words back into a sentence
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
cv=CountVectorizer(binary=True,ngram_range=(1,1))
```

```python
x=cv.fit_transform(corpus).toarray()
```
